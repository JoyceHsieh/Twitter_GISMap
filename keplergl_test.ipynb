```python
import pandas as pd
import keplergl
from keplergl import KeplerGl
import h3
import requests
import geopandas as gpd
from pandasql import PandaSQL
pdsql = PandaSQL()

import arcgis
from arcgis.gis import GIS
from arcgis.mapping import WebMap, WebScene
from arcgis.features import FeatureLayer
from arcgis.features import GeoAccessor, GeoSeriesAccessor
```

```python
# https://developers.arcgis.com/rest/services-reference/query-feature-service-layer-.htm
# This query will return all records (where=1=1), all fields (outFields=*), the correct projection for H3 (outSR=4326), and geojson (f=geojson)
# You could modify this query to return only the records you want, e.g., where Polarity='neg' AND Emotion='sad' rather then the SQL query above

url = requests.get("https://services.arcgis.com/q3Zg9ERurv23iysr/ArcGIS/rest/services/tweets_for_testing_02082022/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=geojson")
df = gpd.GeoDataFrame().from_features(url.json())
df
```

```python
# Adding h3 hex id
def calculate_h3_7(lat, lng):
    h3_hex = h3.geo_to_h3(lat, lng, resolution_7)
    return h3_hex
resolution_7 = 7

def calculate_h3_8(lat, lng):
    h3_hex = h3.geo_to_h3(lat, lng, resolution_8)
    return h3_hex
resolution_8 = 8
```

```python
df['h3_7'] = df.apply(lambda row: calculate_h3_7(row['lat'], row['lon']), axis=1)
df['h3_8'] = df.apply(lambda row: calculate_h3_8(row['lat'], row['lon']), axis=1)
df
```

```python
# Creating a df for the data we want to analyze
# FYI, if you include the geometry column above, you will receive an error - assumption is that that column would not exist in the future only lat/lon values
df_for_sql = df[['OBJECTID','Polarity', 'Emotion','Cognition','created_at_Converted','h3_7']]
df_for_sql
```

```python
# Counting number of tweets in a H3 location using SQL
pd.set_option('max_columns', None)
# pd.set_option("max_rows", None)
df2 = pdsql("SELECT COUNT(OBJECTID), Polarity, Emotion, Cognition, h3_7 FROM df_for_sql WHERE Polarity='neg' AND Emotion='sad' GROUP BY h3_7;")
df2
```

```python
# H3-related data - see here: https://agis.maps.arcgis.com/home/item.html?id=95c3e57bd0ea4b0c9968016a440eecff
url = requests.get("https://services.arcgis.com/q3Zg9ERurv23iysr/ArcGIS/rest/services/h3_7_for_testing_02092022/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=geojson")
sdf = gpd.GeoDataFrame().from_features(url.json())
sdf
```

```python
# Table join
sdf2 = pd.merge(df2, sdf, on='h3_7', how='inner')
sdf2
```

```python
sdf2.dtypes
```

```python
# For Kepler, you need to change the geometry type above to object - this code will do that...
sdf3 = sdf2.astype({'geometry':str})
```

```python
sdf3.dtypes
```

```python
config = {
'version': 'v1',
'config': {
    'mapState': {
        'latitude': 33.5,
        'longitude': -117.5,
        'zoom': 7}
    }
}
```

```python
map1 = KeplerGl(height=400, data={'data_1': sdf3}, config=config)
map1
```
